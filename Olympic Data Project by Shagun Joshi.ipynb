**1. Find In which events did Edwin Flack win a medal?**

```python
ed_flack = olympic[(olympic['Athlete']=='FLACK, Edwin')]
ed_flack
```

```python
ed_flack['Event'].value_counts()
```

**2. Which country has won the most men's gold medals in Athletics over the years? Sort the results alphabetically by the player's names.**

```python
gold_med = olympic[(olympic.Medal=='Gold') & (olympic.Gender=="Men") & (olympic.Sport=='Athletics')]
gold_med
```

```python
mens = gold_med.sort_values(by='Athlete')

mens['NOC'].value_counts()[:10]
```

```python
len(mens['NOC'].value_counts())
```

**6. Display the male gold medal winners for the hockey event over the years. List the results starting with the most recent. Show the Olympic City, Edition, Athlete and the country they represent.**

```python
olympic['Event'].value_counts().head(10)
```

```python
mens_gold_medals = olympic.loc[(olympic['Gender']=="Men") & (olympic['Event']=='hockey') & 
            (olympic['Medal']=="Gold"),['City','Edition','Athlete','NOC']]
mens_gold_medals
```

```python
mens_gold_medals.sort_values(by = 'Edition', ascending=False)
```

**7. How many women's have won the gold medal's till 2008 in Olympic game?**

```python
women_gold = olympic[(olympic['Gender']=="Women") & (olympic["Medal"]=="Gold")]
women_gold
```

```python
women_gold.sort_values(by='Athlete')
```

Many women's have won more than one gold medals in the series of Olympic games, so let's get the total number of women's with the help of value_counts() method.

```python
women_gold['Athlete'].drop_duplicates()
```

```python
women_gold.sort_values(by='Athlete', ascending = True)['Athlete'].value_counts()
```
